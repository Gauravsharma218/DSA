**Types of Objects in a python are:**
1. Mutable
2. Immutable

Immutable Objects:

  1.  int
  2.  float
  3.  string
  4.  bool
  5.  tuple

Mutable Objects:

1. List
2. Set
3. Dictionary

**List**

```python
# list should be included in a square brackets
# list can contain duplicate values in it
my_list = [1,2,3,3,4,5]
print(my_list)
type(my_list)
```

```python
#list can have multiple data types items
my_list=[1,2,3,3,'Gaurav','Rahul',True]
print(my_list)
type(my_list)
```

#  List are Mutable --> changeable

```python
#add elements to a list
my_list = [1,2,3,4]
print(my_list)
my_list.append('Gaurav')
print(my_list)
```

```python
#print elements of a list using their index
print(my_list[0])
print(my_list[4])
```

```python
print(len(my_list))
```

```python
# we also can intiate an empty list
list_2 = []
print(list_2)
print(len(list_2))
```

```python
# deleting an element in a list
list_2 = [2,3,4.8,'English',True,6]
print(list_2)
del list_2[3]
```

```python
#join of two list
list_3 = my_list + list_2
print(list_3)
```

```python
#join of two list
list_1 = [1,2,3,4,5]
list_2 = [6,7,8,9,10]
list_3 = list_1 + list_2
print(list_3)
```

**tuple**

```python
#tuple are introducing using round brackets
#tuples allows multiple data types
tu1 = (1,2,4,5,5)
print(tu1)
```

```python
# converting a list into tuple
tu2 = tuple(my_list)
print(tu2)
```

```python
print(tu2[2])
```

```python
print(tu2.append(5))
```

```python
print(len(tu2))
```

Set --> Curly brackets

```python
# set does not contain duplicate values
my_set = {2,3,5,6,6}
print(my_set)
print(type(my_set))
```

```python
print(my_set[0])
```

```python
# converting a list to a Set
list_5 = {2,3,4,5}
s1 = set(list_5)
print(s1)
```

**Dictionary**

key value pair

```python
my_dict ={
    'name' : 'Gaurav',
    'age'  :  21,
    'country' : 'India'
}
print(my_dict)
type(my_dict)
```

```python
#dictionary does not allow duplicate values
dictionary_2 = {
    'name' : 'Gaurav',
    'age'  :  21,
    'country' : 'India',
    'name' : 'Gaurav',
    'age'  :  21,
    'country' : 'India'
}
print(dictionary_2)
```
